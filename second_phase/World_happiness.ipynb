```python
X = df.drop(columns = ['Happiness Score'])
Y = df['Happiness Score']
```

```python
X
```

```python
Y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.33, random_state=43)
print('Training feature matrix size:',X_train.shape)
print('Training target vector size:',Y_train.shape)
print('Test feature matrix size:',X_test.shape)
print('Test target vector size:',Y_test.shape)
```

```python
from sklearn.metrics import r2_score
maxr2_score = 0
maxrs = 0
for i in range(1,200):
    X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.33, random_state=i)
    lr = LinearRegression()
    lr.fit(X_train,Y_train)
    y_pre = lr.predict(X_test)
    r2 = r2_score(Y_test,y_pre)
    if r2>maxr2_score:
        maxr2_score=r2
        maxrs=i
print('Best accuracy:', maxr2_score, 'Random_state:', maxrs)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.33, random_state = 8)
lr = LinearRegression()
lr.fit(X_train,Y_train)
lr.score(X_train,Y_train)
pred = lr.predict(X_test)
print("Predict value: ", pred)
print('\n')
print("Actual value: ", Y_test)
```

```python
print("MAE: ", mean_absolute_error(Y_test,pred))
print("MSE: ", mean_squared_error(Y_test,pred))
print("RMSE: ", np.sqrt(mean_squared_error(Y_test,pred)))
print("\n")

from sklearn.metrics import r2_score
print("R2_score: ", r2_score(Y_test,pred))
```

```python
rf = RandomForestRegressor(n_estimators = 100)
rf.fit(X_train, Y_train)
rf.score(X_train,Y_train)
pred_ = rf.predict(X_test)
print(rf, ':')
print("MAE: ", mean_absolute_error(Y_test,pred_))
print("MSE: ", mean_squared_error(Y_test,pred_))
print("RMSE: ", np.sqrt(mean_squared_error(Y_test,pred_)))
print("\n")

print("R2_score: ", r2_score(Y_test,pred_))
```

```python
svr = SVR()
svr.fit(X_train, Y_train)
svr.score(X_train,Y_train)
pred_ = svr.predict(X_test)
print(svr, ':')
print("MAE: ", mean_absolute_error(Y_test,pred_))
print("MSE: ", mean_squared_error(Y_test,pred_))
print("RMSE: ", np.sqrt(mean_squared_error(Y_test,pred_)))
print("\n")

print("R2_score: ", r2_score(Y_test,pred_))
```

```python
dtc = DecisionTreeRegressor()
dtc.fit(X_train, Y_train)
dtc.score(X_train,Y_train)
pred_ = dtc.predict(X_test)
print(dtc, ':')
print("MAE: ", mean_absolute_error(Y_test,pred_))
print("MSE: ", mean_squared_error(Y_test,pred_))
print("RMSE: ", np.sqrt(mean_squared_error(Y_test,pred_)))
print("\n")

print("R2_score: ", r2_score(Y_test,pred_))
```

```python
lr = LinearRegression()
lr.fit(X_train, Y_train)
lr.score(X_train,Y_train)
pred_ = lr.predict(X_test)
print(lr, ':')
print("MAE: ", mean_absolute_error(Y_test,pred_))
print("MSE: ", mean_squared_error(Y_test,pred_))
print("RMSE: ", np.sqrt(mean_squared_error(Y_test,pred_)))
print("\n")

print("R2_score: ", r2_score(Y_test,pred_))
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(lr, X, Y, cv=5)
print("Cross Validation : ")
print("Score: ", score)
print("Mean : ", score.mean())
print("Standard Deviation: ", score.std())
```
