```python
rows_drop = row[row>33].index
df.drop(rows_drop, inplace=True)
```

```python
X = df1.drop('CGPA', axis=1)
Y = df1['CGPA']
```

```python
def fillna_with_modes(row):
    modes = row.mode()
    if len(modes) > 0:
        return row.fillna(modes[0])
    else:
        return row

X = X.apply(fillna_with_modes, axis=1)
X
```

```python
from sklearn.preprocessing import LabelEncoder
for col in X.columns:
    label = LabelEncoder()
    X[col] = label.fit_transform(X[col])
```

```python
X.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
from sklearn.model_selection import train_test_split
```

```python
maxr2_score = 0
maxrs = 0
for i in range(1,200):
    X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.33, random_state=i)
    lr = LinearRegression()
    lr.fit(X_train,Y_train)
    y_pre = lr.predict(X_test)
    r2 = r2_score(Y_test,y_pre)
    if r2>maxr2_score:
        maxr2_score=r2
        maxrs=i
print('Best accuracy:', maxr2_score, 'Random_state:', maxrs)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.33, random_state = 20)
lr = LinearRegression()
lr.fit(X_train,Y_train)
lr.score(X_train,Y_train)
pred = lr.predict(X_test)
print("MAE: ", mean_absolute_error(Y_test,pred))
print("MSE: ", mean_squared_error(Y_test,pred))
print("RMSE: ", np.sqrt(mean_squared_error(Y_test,pred)))
print("\n")

from sklearn.metrics import r2_score
print("R2_score: ", r2_score(Y_test,pred))
```

```python
dt = DecisionTreeRegressor()
dt.fit(X_train,Y_train)
dt.score(X_train,Y_train)
pred = dt.predict(X_test)
print("MAE: ", mean_absolute_error(Y_test,pred))
print("MSE: ", mean_squared_error(Y_test,pred))
print("RMSE: ", np.sqrt(mean_squared_error(Y_test,pred)))
print("\n")

print("R2_score: ", r2_score(Y_test,pred))
```

```python
from sklearn.model_selection import GridSearchCV

para = {'criterion':['squared_error','absolute_error'],
        'n_estimators':[100,200,300,400,500]
       }
rf = RandomForestRegressor()
cl = GridSearchCV(rf,para)
cl.fit(X_train,Y_train)

print(cl.best_params_)
```

```python
rf = RandomForestRegressor(criterion = 'squared_error', n_estimators = 200)
rf.fit(X_train, Y_train)
rf.score(X_train,Y_train)
pred_ = rf.predict(X_test)
print(rf, ':')
print("MAE: ", mean_absolute_error(Y_test,pred_))
print("MSE: ", mean_squared_error(Y_test,pred_))
print("RMSE: ", np.sqrt(mean_squared_error(Y_test,pred_)))
print("\n")

print("R2_score: ", r2_score(Y_test,pred_))
```

```python
knn = KNeighborsRegressor()
knn.fit(X_train, Y_train)
knn.score(X_train,Y_train)
pred_ = knn.predict(X_test)
print(knn, ':')
print("MAE: ", mean_absolute_error(Y_test,pred_))
print("MSE: ", mean_squared_error(Y_test,pred_))
print("RMSE: ", np.sqrt(mean_squared_error(Y_test,pred_)))
print("\n")

print("R2_score: ", r2_score(Y_test,pred_))
```

```python
svr = SVR()
svr.fit(X_train, Y_train)
svr.score(X_train,Y_train)
pred_ = svr.predict(X_test)
print(svr, ':')
print("MAE: ", mean_absolute_error(Y_test,pred_))
print("MSE: ", mean_squared_error(Y_test,pred_))
print("RMSE: ", np.sqrt(mean_squared_error(Y_test,pred_)))
print("\n")

print("R2_score: ", r2_score(Y_test,pred_))
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(lr, X, Y, cv=5)
print("Cross Validation : ")
print("Score: ", score)
print("Mean : ", score.mean())
print("Standard Deviation: ", score.std())
```
